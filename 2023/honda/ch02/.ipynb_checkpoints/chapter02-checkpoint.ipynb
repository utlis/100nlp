### 16. ファイルをN分割する
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．

````python
df = pd.read_table("popular-names.txt", header=None, names = ['name', 'sex', 'population', 'year'])
df
````

````python
def split(file, N):
    file_name = 'file_split'
    with open(file, 'r') as f:
        count = sum([1 for line in f]) # ファイルの行数
        for line in f:
            for i in range(N):
                file_name += str(i) # ファイル名の取得
                with open('file_name', 'w') as g:
                    while g <= count/N:
                        g.write(line)
                        print(line)
````

### 17. １列目の文字列の異なり
1列目の文字列の種類（異なる文字列の集合）を求めよ．確認にはcut, sort, uniqコマンドを用いよ．

````python
df = pd.read_table('popular-names.txt', header=None)
df
types = 0
dic = {}
for name in df.iloc[:,0]:
    if dic.get(name) != None:
        dic[name] += 1
    else:
        dic[name] = 1
        types += 1
print(types)
print(dic)
````

````python
with open('popular-names.txt', 'r') as file:
    distinct_strings = set(line.split()[0] for line in file)
print(distinct_strings)
````

```
cut -f 1 popular-names-1.txt | sort -u | wc  # uniqなし
cut -f 1 popular-names-1.txt | sort | uniq | wc
cut -f 1 popular-names-1.txt | sort | uniq  > popular-names-types.txt # ファイル出力
## 136     136     943 (行数、単語数、バイト数)
```

### 19. 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

````python
df = pd.read_table('popular-names.txt', header=None)
df
# types = 0
dic = {}
for name in df.iloc[:,0]:
    if dic.get(name) != None:
        dic[name] += 1
    else:
        dic[name] = 1
        # types += 1
#  print(types)
print(dic)
````

````python
name_freq = sorted(dic.items(), key=lambda x:x[1], reverse=True) 
# re-order by the value of dic, lambda (anonymous inline function)
name_freq
````

```
cut -f 1 popular-names-1.txt | sort | uniq -c | sort -r  
# re-order colmun 1 alphabetically -> delete duplicated lines and record the frequencies of each line -> re-order the lines in descending order by the frequencies
cut -f 1 popular-names-1.txt | sort | uniq -c | sort -r > popular-names-freq.txt # ファイル出力
```
