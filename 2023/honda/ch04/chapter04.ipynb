### 30. 形態素解析結果の読み込み
形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
text_list = []
sent_list = []
with open('neko.txt.mecab', 'r') as f:
    for line in f:
        if line == 'EOS\n': # the end of the sentence??
            text_list.append(sent_list)
            sent_list = []
        elif line == '\n': # ignore newline
            continue
        else:
            list_1 = line.split('\t') # split surface and other information
            # print(list_1)
            list_2 = list_1[1].split(',') # split information
            sent_list.append({'surface': list_1[0], 'base': list_2[6], 'pos': list_2[0], 'pos1': list_2[1]})
            # print(list_2)
```

### 34. 名詞の連接
名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．

```python
# too much if statements?

noun_phrase = []
words = ''
for sent in text_list:
    for i in range(len(sent)):
    # for morph in sent:
        if i == len(sent): # the end of the sentence
            if words != '':
                noun_phrase.append(words)
                words = '' # update the string
        else: # 0 <= i <= len(sent) - 1
            # if sent[i]['pos'] != '名詞' and words != '':
                # noun_phrase.append(words)
               #  words = '' # update the string
            # elif sent[i]['pos'] != '名詞' and words == '':
            if sent[i]['pos'] != '名詞' :
                continue
            elif sent[i]['pos'] == '名詞':
                if sent[i+1]['pos'] != '名詞' and words == '':
                    continue
                elif sent[i+1]['pos'] != '名詞' and words != '':
                    noun_phrase.append(words)
                    words = '' # update the string
                elif sent[i+1]['pos'] == '名詞' and words == '':
                    words = sent[i]['surface'] + sent[i+1]['surface']
                elif sent[i+1]['pos'] == '名詞' and words != '':
                    words += sent[i+1]['surface']
```

### 35. 単語の出現頻度
文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．

```python
# 記号等の確認
punct = []
for sent in text_list:
    for morph in sent:
        if morph['pos'] == '記号':
            if morph['base'] not in punct:
                punct.append(morph['base'])
punct
```

```python
# key=word, value=frequency

freq_dic = {}
for sent in text_list:
    for morph in sent:
        #if  morph['pos'] == '記号':
        if morph['base'] in punct: 
            continue
        elif morph['base'] not in freq_dic:
            freq_dic[morph['base']] = 1
        else:
            freq_dic[morph['base']] += 1
            
```

```python
freq_dic 
# del freq_dic['']
# del freq_dic['\u3000']
```

```python
# sort in descending order of the freqency
freq_dic_1 = sorted(freq_dic.items(), key=lambda x:x[1], reverse=True)
freq_dic_1
```

### 37. 「猫」と共起頻度の高い上位10語
「猫」とよく共起する（共起頻度が高い）10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．

```python
# 　文内での「猫」との共起頻度を計算、文中の「猫」の頻度は問わない場合
# key=word, value=frequency
neko_dic = {}
for sent in text_list:
    surfaces = []
    for morph in sent:
        surfaces.append(morph['base'])
    if '猫' not in surfaces:
        continue
    else:
        for morph in sent:
            #if  morph['pos'] == '記号':
            if morph['base'] in punct: 
                continue
            elif morph['base'] in neko_dic:
                neko_dic[morph['base']] += 1
            else:
                neko_dic[morph['base']] = 1
del neko_dic['猫']
```

```python
neko_dic
neko_dic_1 = sorted(neko_dic.items(), key=lambda x:x[1], reverse=True)
neko_dic_1
```

```python
# 　文内での「猫」との共起頻度を計算、文中の「猫」の頻度も考慮する場合
# key=word, value=frequency
neko_dic_2 = {}
for sent in text_list:
    surfaces = []
    for morph in sent:
        surfaces.append(morph['base'])
    if '猫' not in surfaces:
        continue
    else:
        neko = surfaces.count('猫') # the frequency of 「猫」 in the sentence
        for morph in sent:
            #if  morph['pos'] == '記号':
            if morph['base'] in punct: 
                continue
            elif morph['base'] in neko_dic_2:
                neko_dic_2[morph['base']] += neko
            else:
                neko_dic_2[morph['base']] = neko
del neko_dic_2['猫']
```

```python
neko_dic_2
neko_dic_3 = sorted(neko_dic.items(), key=lambda x:x[1], reverse=True)
neko_dic_3
```
