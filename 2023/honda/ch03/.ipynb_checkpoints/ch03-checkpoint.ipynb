### 25. テンプレートの抽出
記事中に含まれる「基礎情報」テンプレートのフィールド名と値を抽出し，辞書オブジェクトとして格納せよ．

```python
with open('jawiki-uk.txt', 'r') as f:
    file = f.read()
re.findall(r'{{基礎情報.*?$(.*?)^}}', file, re.MULTILINE + re.DOTALL) # ^: 文字列の先頭と各改行の直後
# $: 文字列の末尾、文字列の末尾の改行の直前、改行の前にマッチ
```

```python
text = re.findall(r'{{基礎情報.*?$(.*?)^}}', file, re.MULTILINE + re.DOTALL)
re.findall(r'\|(.*?)\s*=\s*(.*?)(?:(?=\n\|)|(?=\n$))' , text[0], re.MULTILINE + re.DOTALL) # (?:): グループ化、後から参照できない、| : or、?=: 先読み
# re.findall(r'\|(.*?)\s*=\s*(.*?)(?:(?=\n\|)|(?=\n$))' , text[0], re.MULTILINE + re.DOTALL)
```

```python
import re
re_exp_1 = r'{{基礎情報.*?$(.*?)^}}'
re_exp_2 = r'\|(.*?)\s*=\s*(.*?)(?:(?=\n\|)|(?=\n$))' 

import re
with open('jawiki-uk.txt', 'r') as f:
    file = f.read()
file_list = re.findall(re_exp_1, file, re.MULTILINE + re.DOTALL) 
file_list = re.findall(re_exp_2, file_list[0], re.MULTILINE + re.DOTALL) # re.DOTALL: '.' 特殊文字を、改行を含むあらゆる文字にマッチさせる
file_list 

dic = dict(file_list)
dic
# リストの下処理
# file_list = file_list[0].split('\n|')
# file_list.remove('')
# file_list
# file_list[-1] = file_list[-1][:-1]
# file_list

# file_dic = {}
# for item in file_list:
#     file_dic[]

```

### 26. 強調マークアップの除去
25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ（参考: マークアップ早見表）．

```python
def delete_markup_1(value):
    re_exp = r'\'{2,5}'
    return re.sub(re_exp, '', value)

for k, v in dic.items():
    dic[k] = delete_markup_1(v)
    # print(k, v)
dic
```

### 27. 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）．

```python
re.sub(r'\[\[([^|]*?\|)??([^|]*?)\]\]', r'\2', '確立形態3 : [[グレートブリテン及びアイルランド連合王国]]成立<br />（[[合同法 (1800年)|1800年合同法]]）')

```

```python
## [[記事名]]、[[記事名|表示文字]]、[[記事名#節名|表示文字]]
def delete_markup_2(value):
    re_exp = r'\'{2,5}'
    text = re.sub(re_exp, '', value)
    re_exp_1 = r'\[\[([^|]*?\|)??([^|]*?)\]\]' # ??：直前の文字が０個か1個の場合にマッチ／最短一致（条件に合う最短の部分に一致）
    # re_exp_1 = r'\[\[([^|]*?\|)??([^|]*?)\]\]'
    return re.sub(re_exp_1, r'\2', text) # re.sub(re_exp_1, r'\2', text) # \number: 同じ番号のグループの中身にマッチ

for k, v in dic.items():
    dic[k] = delete_markup_2(v)
    # print(k, v)
dic #国家、国章画像のリンクが削除できていない
```

### 28. MediaWikiマークアップの除去
27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．

```python
def delete_markup_3(value):
    re_exp = r'\'{2,5}'
    text = re.sub(re_exp, '', value)
    
    re_exp_1 = r'\[\[([^|]*?\|)??([^|]*?)\]\]'
    text = re.sub(re_exp_1, r'\2', text)
    
    #外部リンク
    re_exp_3 = r'https?://[\w!\?/\+\\-~=;\.,*&@#$%\|\(\)\'\[\]]+'
    re_exp_4 = r'url\='
    text = re.sub(re_exp_3, '', text)
    text = re.sub(re_exp_4, '', text)  
    
    # テンプレートの削除 -> '他元首等氏名2'を要修正
    re_exp_5 = r'(\{\{)|(\}\})'
    text = re.sub(re_exp_5, ' ', text)
    return text

for k, v in dic.items():
    dic[k] = delete_markup_3(v)
    # print(k, v)
dic
```

### 29. 国旗画像のURLを取得する
テンプレートの内容を利用し，国旗画像のURLを取得せよ．（ヒント: MediaWiki APIのimageinfoを呼び出して，ファイル参照をURLに変換すればよい）

```python
# https://amaru-ai.com/entry/2022/10/06/084919#29-%E5%9B%BD%E6%97%97%E7%94%BB%E5%83%8F%E3%81%AEURL%E3%82%92%E5%8F%96%E5%BE%97%E3%81%99%E3%82%8B のまま

import requests
#HTTPライブラリ

def get_url(text):
    url_file = text['国旗画像'].replace(' ', '_')
    url = 'https://commons.wikimedia.org/w/api.php?action=query&titles=File:' + url_file+'&prop=imageinfo&iiprop=url&format=json'
    data = requests.get(url)
    return re.search(r'"url":"(.+?)"', data.text).group(1)

print(get_url(dic))
```
